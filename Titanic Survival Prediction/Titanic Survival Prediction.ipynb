```python
# checking for null values

data.isnull().sum()
```

```python
# Drop the 'cabin' column

data = data.drop('Cabin', axis=1)
```

```python
# Fill null values in the 'Age' column with the mean value

mean_age = data['Age'].mean()
data['Age'].fillna(mean_age, inplace=True)
```

```python
data['Age'].dtype
```

```python
# Find the most common value (mode) in the 'Embarked' column

data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
# describe of data in dataframe

data.describe()
```

```python
# Encode 'Survived' column
data['Survived'] = data['Survived'].map({0: 'Not Survived', 1: 'Survived'})

```

```python
data
```

```python
# Visualize the survival distribution

ax = sns.countplot(x='Survived', data=data, palette={'Not Survived': 'gray', 'Survived': 'green'})
sns.set(rc={'figure.figsize': (7, 6)})

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Survival Distribution')
plt.show()


```

```python
# Plotting the survival rate for each passenger class

ax = sns.countplot(x='Pclass', hue='Survived', data=data,palette={0: "gray", 1: "green"})

# Adding labels to the bars

for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting the survival rate by Age category

# Create Age categories

data['Age_range'] = pd.cut(data['Age'], bins=[0, 18, 35, 60, 100], labels=['0-18', '19-35', '36-60', '61+'])

# Plotting the survival distribution based on Age categories
plt.figure(figsize=(10,7))
ax = sns.countplot(x='Age_range', hue='Survived', data=data, palette={0: "gray", 1: "green"})

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Survival by Age Category')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting the Age category and Gender 

ax = sns.countplot(data = data, x = 'Age_range', hue = 'Sex',palette =  {'male':'purple','female':'pink'}) 
sns.set(rc={'figure.figsize':(10,7)})

# Adding labels to the bars

for bars in ax.containers:
    ax.bar_label(bars)
```
